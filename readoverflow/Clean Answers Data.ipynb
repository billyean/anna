```python
import pandas as pd
import numpy as np
```

```python
answers_data = pd.read_csv("posts_answers.csv", usecols=[0,1,2,3,4,5,6])
answers_data = answers_data.iloc[:180,:]
answers_data
```

```python
users_data = pd.read_csv("users.csv", usecols=[0,1,2,3,4])
users_data
```

```python
answers_joined = answers_data.join(users_data.set_index("id"), on="owner_user_id")
```

```python
answers_joined = answers_joined[answers_joined['up_votes'].notnull()]
answers_joined
```

```python
posts_data = pd.read_csv("stackoverflow_posts.csv", usecols=[0,1,2,3,4])
posts_data
```

```python
answers_joined = answers_joined.join(posts_data.set_index("id"), on="parent_id", lsuffix='_answers', rsuffix='_posts')
answers_joined
```

```python
answers_joined.to_csv('answers_joined.csv')
```
